```python
import cv2,time
```

```python
video = cv2.VideoCapture(0)

time.sleep(3)
a = 1
while True:
    a = a+1
    check,frame = video.read()
    print(frame)
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    cv2.imshow("CLIP IS CAPTURED ", gray)
    key =  cv2.waitKey(1)
    if key == ord('q'):
        break




video.release()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)
a = 1

while True:
    a = a + 1
    check,frame = cap.read()
    print(frame)
    gray = cv2.cvtColor(frame,cv2.COLOR_RGB2GRAY)
    cv2.imshow("VIDEO IS CAPTURING ",gray)
    blur = cv2.GaussianBlur(gray,(21,21),0)
    cv2.imshow("VIDEO IS CAPTURING ",blur)
    key = cv2.waitKey(1)
    if key == ord('q'):
        break
print(a)
cv2.destroyAllWindows()
cap.release()

    
```
