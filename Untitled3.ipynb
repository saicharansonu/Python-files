```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv(r"C:\Users\G.VENKATARAMANA\ML datasets\Indian Liver Patient Analysis\Indian Liver Patient Dataset (ILPD).csv")
```

```python
data.describe()
```

```python
data=data.replace({'Male':0,'Female':1})
```

```python
data.isnull()
```

```python
data=data.bfill(axis=0) # axis=0 represents rows
data
```

```python
data.isnull().sum()
```

```python
data.head(10)
```

```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
x.head(10)
```

```python
y.head(10)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=7)
```

```python
sc_X = StandardScaler()
x_train = sc_X.fit_transform(x_train)
x_test = sc_X.transform(x_test)
```

```python
regressor = LinearRegression()
regressor.fit(x_train,y_train)
y_pred = regressor.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```
